Revenue Over Time

```python
import plotly.express as px

fig1 = px.line(retail_df.groupby('Date')['Revenue'].sum().reset_index(),
               x='Date', y='Revenue', title='Revenue Over Time')

fig1.show()
```

Category Revenue

```python
cat_rev = retail_df.groupby('Category')['Revenue'].sum().reset_index()

fig2 = px.bar(cat_rev, x='Category', y='Revenue', title='Revenue by Category',
              color='Category')

fig2.show()
```

Top 10 Products

```python
top_products = (retail_df.groupby('Product')['Revenue'].sum()
                .sort_values(ascending=False)
                .head(10)
                .reset_index())

fig3 = px.bar(top_products, x='Product', y='Revenue',
              title='Top 10 Products by Revenue')

fig3.show()
```

Revenue by Region

```python
region_rev = retail_df.groupby('Region')['Revenue'].sum().reset_index()

fig4 = px.bar(region_rev, x='Region', y='Revenue', title='Revenue by Region',
              color='Region')

fig4.show()
```

KPI cards

```python
total_revenue = retail_df['Revenue'].sum()
avg_order_value = retail_df['Revenue'].mean()
total_quantity = retail_df['Quantity'].sum()

print("Total Revenue:", total_revenue)
print("Average Order Value:", avg_order_value)
print("Total Quantity Sold:", total_quantity)
```
